Create a function that adds two numbers together and then determines if the number is even or odd. Copy and paste the function definition below.

```python
def is_even(number_1, number_2):
    added_numbers = number_1 + number_2

    if added_numbers % 2 == 0:
        return True
    else:
        return False
```

"A leap year is a year that is divisible by 4 except for years that are evenly divisible by 100, although years divisible by 400 are leap years".

Write a function using Python that accepts a single parameter (year) and determines if the year is a leap year or not. Copy and paste the function definition below.


```python
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False
```



Use the function previously created to answer the question below:

 

Will the year 2032 be a leap year?


```python
is_leap_year(2032)
```



The Chinese Zodiac is determined by birth year, with 12 different animals representing different years of birth in a cyclical manner. The order goes as follows:
(1900) Rat       0
(1901) Ox        1
(1902) Tiger     2
(1903) Rabbit    3
(1904) Dragon    4
(1905) Snake     5
(1906) Horse     6
(1907) Sheep     7
(1908) Monkey    8
(1909) Rooster   9
(1910) Dog       10
(1911) Pig       11
The zodiac sign for any given year can be calculated by subtracting 1900 from the year and then getting its remainder when divided by 12. For example, the year 2022 would be the year of the tiger because 2022 - 1900 = 122 and 122 % 12 = 2. The number 2 (as seen above) corresponds to the tiger.
Create a function that accepts one parameter (year) and uses multiple if-statements to print out the zodiac sign depending on the year given. Copy and paste the function definition below.


```python
def chinese_zodiac(year):
    zodiac_number = (year - 1900) % 12
    
    if zodiac_number == 0:
        return 'Rat'
    elif zodiac_number == 1:
        return 'Ox'
    elif zodiac_number == 2:
        return 'Tiger'
    elif zodiac_number == 3:
        return 'Rabbit'
    elif zodiac_number == 4:
        return 'Dragon'
    elif zodiac_number == 5:
        return 'Snake'
    elif zodiac_number == 6:
        return 'Horse'
    elif zodiac_number == 7:
        return 'Sheep'
    elif zodiac_number == 8:
        return 'Monkey'
    elif zodiac_number == 9:
        return 'Rooster'
    elif zodiac_number == 10:
        return 'Dog'
    else:
        return 'Pig'
```



Use the function previously written to answer the question below.

 

The year 2073 will be associated with which zodiac sign?


```python
chinese_zodiac(2073)
```
